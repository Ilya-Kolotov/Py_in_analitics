### Задача 1.1.

```python
import pandas as pd
df = pd.read_csv('kc_house_data.csv')
df.head()
```

## Задача 1.2
Изучите стоимости недвижимости

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['price'], edgecolor = 'black', color = '#AA0000')
plt.title('Распределение стоимости')
plt.xlabel('Цена')
plt.ylabel('Количество');
```
